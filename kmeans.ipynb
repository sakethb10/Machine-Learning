```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("wine-clustering.csv")
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.cluster import KMeans
```

```python
selected_features=df[['OD280','Alcohol']]
model=KMeans(n_clusters=3, random_state=42)
model.fit(selected_features)
```

```python
y_pred=model.fit_predict(selected_features)
y_pred
```

```python
centers=model.cluster_centers_
print(centers)
```

```python
sns.scatterplot(x=selected_features['OD280'], y=selected_features['Alcohol'], hue=model.labels_)
plt.scatter(centers[:,0], centers[:,1], s=200, c='red')
```
